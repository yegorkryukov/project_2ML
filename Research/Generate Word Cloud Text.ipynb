```python
def GetListOfWords(text):

    # Source: https://machinelearningmastery.com/clean-text-machine-learning-python/
    
    # split into words
    tokens = word_tokenize(text)

    # convert to lower case
    tokens = [w.lower() for w in tokens]

    # remove punctuation from each word
    table = str.maketrans('', '', string.punctuation)
    stripped = [w.translate(table) for w in tokens]

    # remove remaining tokens that are not alphabetic
    words = [word for word in stripped if word.isalpha()]

    # filter out stop words (i.e., it, is, the, etc.)
    from nltk.corpus import stopwords
    stop_words = set(stopwords.words('english'))
    words = [w for w in words if not w in stop_words]
    return(words)
```

```python
# Record starting time
start = time.time()

db_list = ["bbc_db"]
            #["reason_db", "slate_db", "tft_db", "washpost_db", "washtimes_db"]
            #"amercons_db", "bbc_db", "breitbart_db", "db_db", "dailywire_db", "foxnews_db", "guardian_db"]
            #"hill_db", "intercept_db", "mj_db", "msnbc_db", "nr_db", "nypost_db", "oann_db", "politico_db"]
           

conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)

for db in db_list:
    
    database = client[db]
    print()
    print(db)
    print('-----------------------')

    # Get all documents from collection
    documents = database.HTML_pages.find()
    
    # Create blank list to store words
    text_list = []

    # Loop through all documents in collection
    for i in range(0,documents.count()):

        # Print out progress tracker
        print(str(i) + "/" + str(documents.count()))

        # Grab the HTML
        html = documents[i]['html']
        
        # Get all paragraph tags from HTML
        paragraphs = re.findall('<p>(.+?)</p>', html)
        
        # Combine paragraphs into one string
        text = " ".join(paragraphs)
        
        # Remove tags and other unwanted text
        text = re.sub('<[^>]+>', '', text)
        text = re.sub('{[^>]+}', '', text)
        text = re.sub('&amp', '', text)
        text = re.sub('&rsquo', '', text)
        text = re.sub('&rdquo', '', text)
        text = re.sub('&ndash', '', text)
        text = re.sub('&ldquo', '', text)
        text = re.sub('&nbsp', ' ', text)
        text = re.sub('&quot;', ' ', text)

        # Run natural language to clean text
        text_list += GetListOfWords(text)
        
    # Turn list of words into string to write to text file
    text_string = ' '.join(text_list)
    
    # Create file name based off of database name
    filename = db + ".txt"

    # Open the file and write the string to it (Must have UTF8 encoding to handle foreign characters)
    with open(filename, "w", encoding='utf8') as text_file:
        text_file.write(text_string)

# Record end time
end = time.time()

# Print out how many minutes elapsed
print((end - start)/60)
```
