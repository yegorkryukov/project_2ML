```python
# Dependencies
import re
import pymongo
from bs4 import BeautifulSoup as bs
import pandas as pd
import time
```

```python
# Initialize PyMongo to work with MongoDBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
def GetLinks(db_name, site_name):
    
    # Print database name
    print()
    print(db_name)
    print('---------------------------')
    
    # Get database from Mongo
    db = client[db_name]
    
    # Get all documents from HTML_pages collection
    documents = db.HTML_pages.find()
    
    # Count the number of documents
    count = documents.count()

    # Create dictionary to count the number of external links
    # This must match the chord diagram order
    links = {'newrepublic.com':0, 'motherjones.com':0, 'slate.com':0, 'theintercept.com':0,
             'thedailybeast.com':0, 'theatlantic.com':0, 'washingtonpost.com':0, 'politico.com':0,
             'theguardian.com':0, 'bbc.com':0, 'breitbart.com':0, 'foxnews.com':0, 
             'nypost.com':0, 'theamericanconservative.com':0, 'washingtontimes.com':0, 'dailywire.com':0,   
             'reason.com':0, 'thehill.com':0, 'thefiscaltimes.com':0, 'oann.com':0
              }

    # Counter used to track progress
    i = 1
    
    # Counter used to track documents that aren't HTML
    non_html = 0

    # Iterate through each document
    for document in documents:

        # Get the HTML item in the document
        html = document['html']
        
        # Test if document contains HTML
        if ("<!DOCTYPE HTML" in html[0:50].upper()):
        
            # Print progress
            #print(str(round((i / count)*100, 2))+"%")
            print(str(i) + "/" + str(count))

            # Convert string to BeautifulSoup object 
            soup = bs(html, 'lxml')

            # Find all anchor tags and loop through each one
            for url in soup.findAll('a'): 

                # Convert url to string
                url = str(url)

                try:
                    # Find where 'www.' and '.com' are in the url string
                    start = url.index("www.") + 4
                    end = url.index(".com") + 4

                    # Splice external link from url
                    ext_link = url[start:end]

                    # Increment dictionary key/value pair
                    links[ext_link] += 1

                # Error handling
                except (KeyError, ValueError):
                    pass
        
        # What to do if document isn't HTML        
        else:
            
            # Increment counter and print out what the non_HTML looks like
            non_html += 1
            #print(str(non_html) + ": " + html[0:25])

        # Increment progress counter
        i += 1
        
    # Print the total number of documents that weren't HTML    
    #print("Non HTML: " + str(non_html))
    
    # Create empty dictionary for scaled external links
    scaled_links = {}
    
    # Create scale variable to standardize number of pages per site
    # For example, if a collection has 5,000 documents, then reduce external links by 20%
    scale = (1000 / count)
    
    # Loop through each key/value pair in the links dictionary
    for key, value in links.items():
        
        # Create a new key/value pair that is "scaled"
        scaled_links[key] = round(value * scale)
    
    # Set own site's count to zero (e.g., {BBC : 0})
    scaled_links[site_name] = 0
        
    print(links)
    print("Scale: ", str(round(scale,2)))
    print(scaled_links)
    
    # Intialize blank list
    link_list = []
    
    # Loop through each key/value in the scaled links dictionary
    for key, value in scaled_links.items():
        
        # Append value to list
        link_list.append(value)
        
    print(link_list)

    # Return results
    return(link_list)
```

```python
start = time.time()

atlantic_links = GetLinks("atlantic_db", "theatlantic.com")
amercons_links = GetLinks("amercons_db", "theamericanconservative.com")
bbc_links = GetLinks("bbc_db", "bbc.com")
breitbart_links = GetLinks("breitbart_db", "breitbart.com")
db_links = GetLinks("db_db", "thedailybeast.com")
dw_links = GetLinks("dailywire_db", "dailywire.com")
foxnews_links = GetLinks("foxnews_db", "foxnews.com")
guardian_links = GetLinks("guardian_db", "theguardian.com")
hill_links = GetLinks("hill_db", "thehill.com")
intercept_links = GetLinks("intercept_db", "theintercept.com")
mj_links = GetLinks("mj_db", "motherjones.com")
newrepublic_links = GetLinks("nr_db", "newrepublic.com")
nypost_links = GetLinks("nypost_db", "nypost.com")
oann_links = GetLinks("oann_db", "oann.com")
politico_links = GetLinks("politico_db", "politico.com")
reason_links = GetLinks("reason_db", "reason.com")
slate_links = GetLinks("slate_db", "slate.com")
tft_links = GetLinks("tft_db", "thefiscaltimes.com")
washpost_links = GetLinks("washpost_db", "washingtonpost.com")       
washtimes_links = GetLinks("washtimes_db", "washingtontimes.com")

end = time.time()
print("Minutes: " + str((end - start)/60))
```

```python
# Make a list of dictionaries to output as JSON
data_json = 
    [{
    "NewRepublic":newrepublic_links, "MotherJones":mj_links, "Slate":slate_links, "Intercept":intercept_links,
    "DailyBeast":db_links, "Atlantic":atlantic_links, "WashPost":washpost_links, "Politico":politico_links,
    "Guardian":guardian_links, "BBC":bbc_links, "Breitbart":breitbart_links, "FoxNews":foxnews_links, 
    "NYPost":nypost_links, "AmerCons":amercons_links, "WashTimes":washtimes_links, "DailyWire": dw_links, 
    "Reason":reason_links,  "Hill":hill_links, "FiscalTimes":tft_links, "OANN":oann_links, 
    }]
```

```python
import json

# Write list of dictionaries to JS file
with open ('data.js', 'w') as outfile:
    json.dump(data_json, outfile)
```
